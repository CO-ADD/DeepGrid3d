```python
#
import os
import numpy as np
```

```python
DataDir = '../Data'
DataFile = 'CS000008317_c1_vp0.npz'

data = np.load(os.path.join(DataDir,DataFile))

grid = data['coords']
channels = data['channels']

print(f"Grid: {grid.shape}")
print(f"Channels: {channels.shape}")
```

```python
import matplotlib.pyplot as plt

# 0 - 30 Angstrom @ 1.5 A spacing
# dim = np.linspace(0, 30, 21)
# X,Y,Z = np.meshgrid(dim,dim,dim)

X = grid[:,0]
Y = grid[:,1]
Z = grid[:,2]

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.scatter3D(X,Y,Z,c=channels[:,3],alpha=0.25, marker='.')
plt.show()
```
